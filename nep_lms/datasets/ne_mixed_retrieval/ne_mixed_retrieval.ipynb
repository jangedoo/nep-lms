```python
# to create a strong testing set, we'll use a modern multi-lingual model as a baseline
# in the test set only those rows will be present whose query-doc similarity is less than max(query to negatives)
# for train and valid, we'll partiton the remaining rows
from sentence_transformers import SentenceTransformer
encoder = SentenceTransformer("intfloat/multilingual-e5-small", device="cuda:1")
```

```python
# they recommend to prefix the query and document when encoding but it is not configured by
# default, so we manually add these
encoder.prompts = {"query": "query: ", "document": "passage: "}
encoder.default_prompt_name = "query"
```

```python
def embed(df, encoder):
    query_embeddings = encoder.encode_query(df['query'].tolist())
    pos_embeddings = encoder.encode_document(df['positive'].tolist())
    neg_embeddings = encoder.encode_document(df['negative_1'].tolist())
    neg2_embeddings = encoder.encode_document(df['negative_2'].tolist())
    neg3_embeddings = encoder.encode_document(df['negative_3'].tolist())
    neg4_embeddings = encoder.encode_document(df['negative_4'].tolist())
    return query_embeddings, pos_embeddings, neg_embeddings, neg2_embeddings, neg3_embeddings, neg4_embeddings

def df_with_pairwise_sim(df, encoder_name:str, query_embeddings, pos_embeddings, neg_embeddings, neg2_embeddings, neg3_embeddings, neg4_embeddings):
    query_pos_sim = encoder.similarity_pairwise(query_embeddings, pos_embeddings)
    query_neg_sim = encoder.similarity_pairwise(query_embeddings, neg_embeddings)
    query_neg2_sim = encoder.similarity_pairwise(query_embeddings, neg2_embeddings)
    query_neg3_sim = encoder.similarity_pairwise(query_embeddings, neg3_embeddings)
    query_neg4_sim = encoder.similarity_pairwise(query_embeddings, neg4_embeddings)
    df[f'{encoder_name}_qp'] = query_pos_sim
    df[f'{encoder_name}_qn1'] = query_neg_sim
    df[f'{encoder_name}_qn2'] = query_neg2_sim
    df[f'{encoder_name}_qn3'] = query_neg3_sim
    df[f'{encoder_name}_qn4'] = query_neg4_sim
    df[f'{encoder_name}_qnmax'] = df[[f'{encoder_name}_qn1', f'{encoder_name}_qn2', f'{encoder_name}_qn3', f'{encoder_name}_qn4']].max(axis=1)
    return df

def populate_df(encoder, encoder_name, df):
    query_embeddings, pos_embeddings, neg_embeddings, neg2_embeddings, neg3_embeddings, neg4_embeddings = embed(df, encoder)
    return df_with_pairwise_sim(df, encoder_name, query_embeddings, pos_embeddings, neg_embeddings, neg2_embeddings, neg3_embeddings, neg4_embeddings)
    
```

```python
_ = populate_df(encoder, 'e5', df)
```

```python
test_df = df.query("e5_qp < e5_qnmax")
test_sample_ids = test_df['sample_id']
train_df = df[~df['sample_id'].isin(test_sample_ids)]
print(f"total rows in test_df {len(test_df):,} and {len(train_df):,} in train_df")

counts = test_df['domain'].value_counts()

# Get the normalized values (percentages)
percentages = test_df['domain'].value_counts(normalize=True)

# Combine them into a DataFrame
test_stats_df = pd.DataFrame({
    'Domain': counts.index,
    'Count': counts.values,
    'Percentage (%)': (percentages * 100).round(2)
})

# Print as markdown
print(test_stats_df.to_markdown(index=False))
```

```python

final_cols = ['query', 'positive', 'negative_1', 'negative_2', 'negative_3', 'negative_4', 'domain']
features = datasets.Features({
    col: datasets.Value("string")
    for col in final_cols
})

ds = datasets.Dataset.from_pandas(train_df[final_cols], preserve_index=False, features=features).train_test_split(test_size=0.1, shuffle=True, seed=100)
ds['valid'] = ds['test']
ds['test'] = datasets.Dataset.from_pandas(test_df[final_cols], preserve_index=False, features=features)
ds
```

```python
ds.push_to_hub("jangedoo/sanoir-general")
```

```python
from sentence_transformers.sentence_transformer.evaluation import RerankingEvaluator

samples = []
negative_cols = [c for c in test_df.columns if c.startswith("negative")]
for _, row in test_df.iterrows():
    negatives = row[negative_cols].tolist()

    samples.append(
        {
            "query": row["query"],
            "positive": [row["positive"]], # important! it should be a list
            "negative": negatives,
        }
    )
```

```python
reranking_evaluator = RerankingEvaluator(
    samples=samples,
    name="pos-vs-4-negs",
    at_k=5,
    batch_size=64,
    show_progress_bar=False
)
rerank_results = reranking_evaluator(encoder)
print(rerank_results)
```
